```python
import numpy
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.mlab as mlab
#import seaborn as sns
%matplotlib inline
```

```python
english_length=[]
eng=[]
with open("text.en", "r") as myfile:
    for line in myfile:
        line=line.strip()
        tokens=line.split()
        english_length.append(len(tokens))
        eng.append(len(tokens))
```

```python
japa_length=[]
jp=[]
with open("text.fr", "r") as myfile:
    for line in myfile:
        line=line.strip()
        tokens=line.split()
        japa_length.append(len(tokens))
        jp.append(len(tokens))
```

```python
fig = plt.figure()
fit = np.polyfit(eng,jp,1)
fit_fn = np.poly1d(fit) 
plt.plot(eng, jp, 'go')
plt.xlabel('Length of English sentences')
plt.ylabel('Length of Japanese sentences')

#plt.axis([0, 6, 0, 20])
plt.show()
fig.savefig('corr.png', dpi=1000)
#data = np.random.multivariate_normal((np.mean(english_length),p.mean(japa_length)), cov, 200)
#df = pd.DataFrame(data, columns=["x", "y"])
#print(eng[1])
#sns.jointplot(eng, jp, kind="reg")
#sns.plt.show()
```

Correlation

```python
sort_ja=sorted(japa_length)
sort_en=sorted(english_length)
N = 5
#men_means = (20, 35, 30, 35, 27)
#men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(0.2+ind, sort_ja[len(sort_ja)-5:], width, color='r')

#women_means = (25, 32, 34, 20, 25)
#women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width+0.2, sort_en[len(sort_en)-5:], width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Length of sentence')
ax.set_title('The high five sentence length')
ax.set_xticks(ind + width+0.2)
ax.set_xticklabels(('1', '2', '3', '4', '5'))

ax.legend((rects1[0], rects2[0]), ('Japanese', 'English'))


def autolabel(rects):

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
fig.savefig('corre.png', dpi=1000)
plt.show()
```

Number of word types

```python
en_types={}
rep=[]
with open("text.en", "r") as myfile:
    for line in myfile:
        line=line.strip()
        tokens=line.split()
        for token in tokens:
            if token in en_types.keys():
                temp=en_types[token]
                en_types[token]=temp+1
            else:
                en_types[token]=1
                
print("number of word types in English:{0}".format(len(en_types.keys())))
counter=0           
for tokens in en_types.keys():
    if en_types[tokens]==1:
        counter+=1
print("number of unique word types in English:{0}".format(counter))
```

```python
jp_types={}
with open("text.fr", "r") as myfile:
    for line in myfile:
        line=line.strip()
        tokens=line.split()
        for token in tokens:
            if token in jp_types.keys():
                temp=jp_types[token]
                jp_types[token]=temp+1
            else:
                jp_types[token]=1

print("number of word types in Japanese:{0}".format(len(jp_types.keys())))
counter=0           
for tokens in jp_types.keys():
    if jp_types[tokens]==1:
        counter+=1
print("number of unique word types in Japanese:{0}".format(counter))
```
