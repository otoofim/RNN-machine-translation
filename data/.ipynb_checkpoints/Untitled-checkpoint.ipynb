```python
import numpy
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
english_length=[]
with open("text.en", "r") as myfile:
    for line in myfile:
        line=line.strip()
        tokens=line.split()
        english_length.append(len(tokens)+(len(tokens)-1))
```

```python
japa_length=[]
with open("text.fr", "r") as myfile:
    for line in myfile:
        line=line.strip()
        tokens=line.split()
        japa_length.append(len(tokens)+(len(tokens)-1))
```

```python
fig = plt.figure()
x = np.mean(english_length) + (np.var(english_length)**(1/2)) * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Length of English sentences')
#plt.ylabel('Probability')
plt.title('Distribution')
plt.legend(loc='best')

#plt.text(9, 0.45, "Mean={0}   Variance={1}".format(np.var(total),np.mean(total)))
plt.title("Mean={1}   Variance={0}".format(np.var(english_length),np.mean(english_length)))
y = mlab.normpdf( bins, np.mean(english_length), (np.var(english_length)**(1/2)))
l = plt.plot(bins, y, 'r', linewidth=1)

#plt.axis([0, , 0, 0.5])
plt.show()
fig.savefig('english.png', dpi=1000)
```

```python
fig = plt.figure()
x = np.mean(japa_length) + (np.var(japa_length)**(1/2)) * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Length of Japanese sentences')
#plt.ylabel('Probability')
plt.title('Distribution')
plt.legend(loc='best')

#plt.text(9, 0.45, "Mean={0}   Variance={1}".format(np.var(total),np.mean(total)))
plt.title("Mean={1}   Variance={0}".format(np.var(japa_length),np.mean(japa_length)))
y = mlab.normpdf( bins, np.mean(japa_length), (np.var(japa_length)**(1/2)))
l = plt.plot(bins, y, 'r', linewidth=1)

#plt.axis([0, , 0, 0.5])
plt.show()
fig.savefig('Japa.png', dpi=1000)
```
